## Using OpenAI's API to correct the raw paragraphs

```python
corrected_passages = []
c = -1

client = OpenAI()

for i in raw_passages['Passages']:
  completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
          {"role": "system", "content": "So I am going to give you a task, I have a list of paragraphs from a book, but they are not in the correct format, I have simply used one python script to split them into paragraphs, but there are many inconsistencies in them, like some words chapter, pg, chapter names, dates, etc are coming. So you have to remove them also at many places words are mix matched like 'th ebright', it should be 'th ebright'. I am going to give you an example , Passage: 'Ch ap te rone ahmadnagar fort twenty months ah ma dn ag ar fort, 13th april It is more than twenty months since we we re brought he re , more than twenty months of my ninth term of imprisonment. Th enew moon, ashimmering crescent in the darkening sky, greeted us on our arrival here. th ebright fortnight of the waxing moon ha dbegun. ever since then each coming of the new moon has been areminder to me that another mo nt hof my imprisonment is over.' Corrected version: 'It is more than twenty months since we were brought here , more than twenty months of my ninth term of imprisonment. The new moon, ashimmering crescent in the darkening sky, greeted us on our arrival here. the bright fortnight of the waxing moon had begun. ever since then each coming of the new moon has been a reminder to me that another month of my imprisonment is over.' So I will give you paragraphs and you just have to make them correct."},
           {"role": "user", "content": i}
          ]
      )
  corrected_passages.append(completion.choices[0].message.content)

  c = c + 1
  if c%100 ==0:
    print(c, "paragraphs, successfully corrected")
  else:
    continue
```

## Creating a dataframe of the passages

```python
corrected_passages_df = pd.DataFrame(corrected_passages)
```

## Generating questions


```python
corrected_passages = pd.read_csv("corrected_passages.csv")
```

```python
corrected_passages
```

```python
generated_questions = []
c = -1

client = OpenAI()

for i in range(len(corrected_passages)):
  completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
          {"role": "system", "content": "So I will give a paragraphs and you have to make the question out of it, so that paragraph would be an answer to your generated question"},
           {"role": "user", "content": corrected_passages.iloc[i, 0]}
          ]
      )
  generated_questions.append(completion.choices[0].message.content)

  c = c + 1
  if c%100 ==0:
    print(c, "paragraphs, successfully corrected")
  else:
    continue
```

```python
generated_dataaset = pd.DataFrame()
```

```python
generated_dataaset['Questions'] = generated_questions
generated_dataaset['Answers'] = corrected_passages
```
